```python
import pandas as pd
import folium
```

```python
chaps = pd.read_csv("chapters_geo.csv")
```

```python
chaps.head()
```

```python
len(chaps.Location.unique())
```

```python
m = folium.Map(location=[42, -100], zoom_start=4)
```

```python
popup_text = '''
<p>
<b>Chapter:</b> {c}
<br>
<b>Institution:</b> {i}
<br>
<b>Location:</b> {l}
<br>
<b>District:</b> {d}
<br>
<b>Charter Date:</b> {cd}
</p>
'''

color_dict = {"Active": 'blue', 'Inactive': 'orange', 'Investigative Hold': 'green', 'Probation': 'green'}

for i, row in chaps.iterrows():
    
    color = color_dict[row['Status']]
    
    html = popup_text.format(c=row['Chapter'],
                               i=row['Institution'],
                               l=row['Location'],
                               d=row['District'],
                               cd=row['Charter Date'])
    
    folium.Marker(location=[row['Lat'], row['Long']], 
                  popup=folium.Popup(html,max_width=2650),
                  icon=folium.Icon(color=color, icon='music')
                 ).add_to(m)
```
