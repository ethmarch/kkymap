```python
import pandas as pd
import folium
```

```python
chaps = pd.read_csv("chapters_geo.csv")
```

```python
chaps.head()
```

```python
m = folium.Map(location=[42, -100], zoom_start=4)
```

```python
popup_text = '''
Chapter: {c}

Institution: {i}

Location: {l}

District: {d}

Charter Date {cd}
'''

for i, row in chaps.iterrows():
    
    folium.Marker(location=[row['Lat'], row['Long']], 
                  popup=folium.Popup(html=popup_text.format(c=row['Chapter'],
                                                            i=row['Institution'],
                                                            l=row['Location'],
                                                            d=row['District'],
                                                            cd=row['Charter Date']),
                                     max_width=200)
                  #icon=folium.Icon(icon='cloud')
                 ).add_to(m)
```

```python
m.save("index.html")
```
